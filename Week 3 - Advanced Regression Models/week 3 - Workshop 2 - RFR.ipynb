```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_rfr = pd.read_csv('ParisHousing.csv')
```

```python
df_rfr.head()
```

```python
df_rfr.info()
```

```python
df_rfr.describe()
```

```python
df_rfr.isnull().sum()
```

#### Preprocessing for SVR

```python
# plot correlation matrix large
plt.figure(figsize=(12, 8))
sns.heatmap(df_rfr.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
df_rfr.columns
```

```python
df_rfr.drop(['hasYard', 'hasPool', 'floors', 'cityCode', 'cityPartRange', 'numPrevOwners', 'made', 'isNewBuilt', 'hasStormProtector', 'basement', 'attic', 'hasStorageRoom', 'hasGuestRoom'], axis=1, inplace=True)
```

```python
df_rfr.head()
```

#### Preprocessing for SVR

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
```

```python
scale.fit(df_rfr.drop('price', axis=1))
```

```python
scaled_features = scale.transform(df_rfr.drop('price', axis=1))
```

```python
df_feat = pd.DataFrame(scaled_features, columns=df_rfr.columns[:-1])
```

```python
df_feat.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_feat
y = df_rfr['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor()
```

```python
rfr.fit(x_train, y_train)
```

```python
predictions = rfr.predict(x_test)
```

```python
# show the model tree predictions in a plot seborn
plt.figure(figsize=(12, 8))
sns.scatterplot(predictions)
plt.show()
```

```python
sns.distplot((y_test-predictions), bins=50)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
```

```python
print('MSE:', metrics.mean_squared_error(y_test, predictions))
```

```python
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
rfr.score(x_test, y_test)
```

```python
rfr.score(x_train, y_train)
```
